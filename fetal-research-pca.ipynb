```python
X=df.drop(["fetal_health"],axis=1)
y=df["fetal_health"]

col_names = list(X.columns)
s_scaler = preprocessing.MinMaxScaler()
X_df= s_scaler.fit_transform(X)
X_df = pd.DataFrame(X_df, columns=col_names)   
X_df.describe().T
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=0.95)
pca.fit(X_df)
principalComponents=pca.transform(X_df)
```

```python
principalDf = pd.DataFrame(data = principalComponents)
```

```python
plt.figure(figsize=(20,10))
sns.boxenplot(data = principalDf)
plt.xticks(rotation=90)
plt.show()
```

```python
X_train, X_test, y_train,y_test = train_test_split(principalDf,y,test_size=0.2,random_state=0)
```

# Logistic Regression

```python
logreg = LogisticRegression(multi_class='multinomial')
logreg.fit(X_train, y_train)
acc_log = round(logreg.score(X_train, y_train) * 100, 2)
acc_log
```

```python
acc_test_log = round(logreg.score(X_test, y_test) * 100, 2)
acc_test_log
```

```python
 import sklearn
predictions=logreg.predict(X_test)
plt.subplots(figsize=(7,5))
cf_matrix = confusion_matrix(y_test, predictions)
sns.heatmap(cf_matrix,annot = True, annot_kws = {'size':15})

acc=accuracy_score(y_test, predictions, normalize=True, sample_weight=None)
f1 = f1_score(y_test, predictions, labels=None, pos_label=1, average='macro', sample_weight=None, zero_division='warn')
re =recall_score(y_test, predictions,labels=None, pos_label=1, average='macro', sample_weight=None, zero_division='warn')
pre= precision_score(y_test,predictions, labels=None, pos_label=1, average='macro', sample_weight=None, zero_division='warn')
js = jaccard_score(y_test, predictions,  labels=None, pos_label=1, average='macro', sample_weight=None, zero_division='warn')
ks = sklearn.metrics.cohen_kappa_score(y_test, predictions, labels=None, weights=None, sample_weight=None)
print (acc,f1, re,pre,js,ks)
```
